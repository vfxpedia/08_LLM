# 2024-11-01 이후 모든 공지/공고 크롤링

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import csv
import re

# 🌟 초기 페이지 URL 설정
BASE_URL = "https://www.i-sh.co.kr"
LIST_PATH = "/main/lay2/program/S1T294C295/www/brd/m_241/"
LIST_URL = f"{BASE_URL}{LIST_PATH}list.do?multi_itm_seqs=1,2,4,8,16,32,64,128,256,512,1024&rows=10&cpage=1"
VIEW_PATH = f"{LIST_PATH}view.do" # 상세 페이지 경로

# 🌟 요청 파일명으로 변경
OUTPUT_FILE = "sh_notice.csv"
# 🌟 요청 필터 조건 유지: 2024-11-01 이후 데이터만 저장
MIN_DATE = "2024-11-01" 

# 🌟 핵심 선택자
TABLE_SELECTOR = "#listTb tbody tr" 
PAGING_CONTAINER_SELECTOR = "div.page" 
TABLE_BODY_SELECTOR = "#listTb tbody" 

# ▶️ 크롤링 실행 함수
def scrape():
    results = []
    headers = ["번호", "제목", "담당부서", "등록일", "조회수", "URL"] 
    page = 1

    # Selenium 설정
    chrome_options = Options()
    chrome_options.add_argument("--headless") 
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    service = Service()
    driver = webdriver.Chrome(service=service, options=chrome_options)
    
    driver.maximize_window()
    driver.implicitly_wait(5)

    print("🚗 Initializing driver and loading page 1...")
    driver.get(LIST_URL)
    
    time.sleep(3) 
    
    # 초기 페이지 로드 대기 (20초 타임아웃)
    try:
        WebDriverWait(driver, 20).until(
            EC.visibility_of_element_located((By.CSS_SELECTOR, TABLE_BODY_SELECTOR))
        )
        print("✅ 초기 목록 테이블 로드 성공.")
    except:
        print("❌ 초기 페이지 로드 실패. 크롤링을 시작할 수 없습니다.")
        driver.quit()
        return

    while True:
        # 🌟 7페이지 제한 조건 제거: MIN_DATE 조건이 만족할 때까지 무제한 크롤링
            
        print(f"\n📄 Crawling page {page} ...")
        
        # 1. 현재 페이지 번호 strong 태그를 찾아서 다음 페이지 대기 기준으로 사용
        try:
            current_page_strong = driver.find_element(By.XPATH, f"//div[@class='page']//strong[text()='{page}']")
        except:
            if page > 1:
                print("❌ 현재 페이지 번호 strong 태그를 찾을 수 없어 페이지 전환 확인이 불가능합니다.")
            break 
        
        # 2. HTML 파싱 및 데이터 추출
        soup = BeautifulSoup(driver.page_source, "html.parser")
        rows = soup.select(TABLE_SELECTOR)
        
        if not rows:
            print("⚠️ 데이터 없음 (목록 끝에 도달)")
            break
            
        data_filtered_out = False
        for row in rows:
            cols = [c.get_text(strip=True) for c in row.select("td")]
            if len(cols) < 5:
                continue
            
            # URL 추출 로직
            title_anchor = row.select_one("td:nth-child(2) a")
            detail_url = ""
            
            if title_anchor and title_anchor.has_attr('onclick'):
                onclick_value = title_anchor['onclick']
                
                # 상세 보기 함수(goView, getDetailView 등)에서 게시물 번호 추출
                match = re.search(r"(?:goView|getDetailView)\((?:['\"])?\s*(\d+)\s*(?:['\"]?)?,?\s*(?:.*?)?\)", onclick_value)
                
                if match:
                    seq_id = match.group(1)
                    # URL 파라미터 'seq'를 사용하여 상세 페이지 URL 조합
                    detail_url = f"{BASE_URL}{VIEW_PATH}?seq={seq_id}"

            num, title, department_name, reg_date, views = cols[0], cols[1], cols[2], cols[3], cols[4]

            if reg_date < MIN_DATE:
                data_filtered_out = True
                print(f"🚫 {MIN_DATE} 이전 데이터({reg_date})가 발견되어 크롤링을 안전하게 종료합니다.")
                break 

            results.append([num, title, department_name, reg_date, views, detail_url])
        
        if data_filtered_out:
            break # MIN_DATE 필터에 걸리면 크롤링 종료

        # 3. 다음 페이지로 이동 로직 (JavaScript 직접 호출)
        next_page = page + 1
        
        # 🌟 다음 페이지 번호가 있다고 가정하고 JavaScript 함수 호출
        js_call = f"getPaging({next_page}, null);"
        
        # 4. JavaScript 직접 호출 및 대기 실행
        try:
            # 4-1. JavaScript 함수 'getPaging'을 직접 호출하여 페이지 이동
            driver.execute_script(js_call)
            print(f"🚀 JavaScript 호출 성공: {js_call}")
            
            # 4-2. 대기: 새 페이지 번호가 strong 태그로 나타날 때까지 기다림
            WebDriverWait(driver, 15).until(
                EC.presence_of_element_located((By.XPATH, f"//div[@class='page']//strong[text()='{next_page}']"))
            )
            
            page += 1

        except Exception as e:
            # 다음 페이지로 이동을 시도했으나 새 페이지가 로드되지 않으면 = 목록의 끝
            print("🏁 다음 페이지로 이동할 수 없습니다. (목록 끝에 도달했거나 로드 실패)")
            break


    driver.quit()
    
    # CSV 저장
    with open(OUTPUT_FILE, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.writer(f)
        writer.writerow(headers)
        writer.writerows(results)

    print(f"\n🎉 최종 완료! 총 {len(results)}개 저장됨 → {OUTPUT_FILE}")

# scrape() 함수 호출
if __name__ == '__main__':
    scrape()
```

# 다운로드 테스트

```python
import os
import csv
import time
import re
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
from glob import glob
import shutil

# --------------------------
# 🌟 설정 변수
# --------------------------
CSV_FILE = "sh_notice.csv"
DOWNLOAD_DIR = os.path.abspath("./downloads_sh") # 다운로드 경로 설정
WAIT_TIME = 20 # Selenium 최대 대기 시간 (초)
DOWNLOAD_WAIT_TIME = 10 # 다운로드 완료를 위한 대기 시간 (초)

# --------------------------
# 🌟 Selenium 함수
# --------------------------

def setup_driver(download_path):
    """
    Chrome WebDriver 설정 (헤드리스 모드 활성화 및 안정화 옵션 적용)

    Args:
        download_path: PDF 다운로드 저장 경로

    Returns:
        WebDriver 객체
    """
    chrome_options = Options()

    # 다운로드 설정
    prefs = {
        "download.default_directory": download_path,
        "download.prompt_for_download": False,
        "download.directory_upgrade": True,
        "plugins.always_open_pdf_externally": True
    }
    chrome_options.add_experimental_option("prefs", prefs)

    # 🌟 최종 안정화 옵션 및 헤드리스 모드 활성화
    chrome_options.add_argument('--disable-gpu')
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    chrome_options.add_argument("--start-maximized") 
    chrome_options.add_argument('--headless') # 👈 충돌 방지를 위한 핵심 수정
    
    # 드라이버 생성
    try:
        service = Service(ChromeDriverManager().install())
        driver = webdriver.Chrome(service=service, options=chrome_options)
        return driver
    except Exception as e:
        print(f"❌ Chrome 드라이버 설정 오류: {e}")
        print("💡 'pip install selenium webdriver-manager' 명령어를 실행했는지 확인하세요.")
        return None

def download_attachments_from_page(driver, url, date, title, download_path):
    """
    공고 상세 페이지에서 첨부파일 다운로드 및 파일명 변경

    Args:
        driver: WebDriver 객체
        url: 공고 상세 페이지 URL
        date: 등록일 (파일명에 사용)
        title: 제목 (파일명에 사용)
        download_path: 다운로드 저장 경로

    Returns:
        int: 다운로드된 파일 개수
    """
    downloaded_count = 0
    
    # 다운로드 시작 전 파일 목록 저장 함수
    def get_download_files():
        return set(glob(os.path.join(download_path, '*')))
    
    initial_files = get_download_files()

    # 🌟 첨부파일로 간주할 확장자 목록 (hwp, docx 등 추가)
    extensions = ['.pdf', '.xls', '.xlsx', '.txt', '.hwp', '.hwpx', '.doc', '.docx']
    ext_regex = r'\.(' + '|'.join([ext[1:] for ext in extensions]) + r')$' # .pdf, .xls 등

    try:
        print(f"\n➡️ 상세 페이지 접속: {url}")
        driver.get(url)

        WebDriverWait(driver, WAIT_TIME).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'div.pageContainer, #mainform')) 
        )
        
        # 🌟 첨부파일 링크 찾기 (확장자 텍스트를 포함하거나 다운로드 함수를 호출하는 모든 <a> 태그)
        xpath_query = f"//*[self::a][contains(@onclick, 'FileDownLoad') or contains(text(), '{extensions[0]}') or contains(text(), '{extensions[1]}') or contains(text(), '.hwp')]"
        
        file_elements = driver.find_elements(By.XPATH, xpath_query)

        if not file_elements:
            print("⚠️ 첨부파일 링크를 찾을 수 없습니다. (PDF, 엑셀, 텍스트, HWP)")
            return 0

        print(f"🔎 총 {len(file_elements)}개의 다운로드 가능 요소 발견.")
        
        for i, element in enumerate(file_elements):
            try:
                original_filename = element.text.strip()
                onclick_value = element.get_attribute('onclick')
                
                # 1. 파일명에서 확장자 추출 시도
                match = re.search(ext_regex, original_filename, re.IGNORECASE)
                ext = match.group(0).lower() if match else None

                # 2. 파일명이 없다면 onclick 값 확인
                if not ext and onclick_value:
                    # Innorix 다운로드 함수가 확장자를 포함하고 있을 수 있음
                    ext_match_in_onclick = re.search(ext_regex, onclick_value, re.IGNORECASE)
                    if ext_match_in_onclick:
                        ext = ext_match_in_onclick.group(0).lower()
                
                # 3. 확장자가 없으면 건너뜀
                if not ext:
                    continue

                # 4. 새 파일명 조합
                # 제목에서 확장자 제거 (중복 방지)
                clean_title = re.sub(ext_regex, '', original_filename, flags=re.IGNORECASE)
                # 제목에서 불가능한 문자 제거 및 길이 제한
                clean_title = re.sub(r'[\\/:*?"<>|]', '', clean_title)[:50].strip()
                new_filename = f"({date}) {clean_title}{ext}"
                
                print(f"  - 다운로드 시도: '{original_filename}' -> '{new_filename}'")
                
                # 5. 클릭
                driver.execute_script("arguments[0].click();", element)
                
                # 6. 다운로드 완료 대기 및 파일 이름 변경
                time.sleep(DOWNLOAD_WAIT_TIME) 

                current_files = get_download_files()
                newly_downloaded = current_files - initial_files
                
                renamed = False
                for f_path in newly_downloaded:
                    f_name = os.path.basename(f_path)
                    
                    if not f_name.endswith(('.crdownload', '.tmp')):
                        try:
                            final_new_filepath = os.path.join(download_path, new_filename)
                            
                            # 파일명 충돌 방지: (i) 첨가
                            if os.path.exists(final_new_filepath):
                                base, ext_name = os.path.splitext(new_filename)
                                final_new_filepath = os.path.join(download_path, f"{base}_{downloaded_count+1}{ext_name}")
                            
                            os.rename(f_path, final_new_filepath)
                            print(f"  - 파일명 변경 성공: {f_name} -> {os.path.basename(final_new_filepath)}")
                            initial_files.add(final_new_filepath) 
                            initial_files.discard(f_path)
                            downloaded_count += 1
                            renamed = True
                            break
                        except Exception as rename_e:
                            print(f"  - 파일명 변경 오류 발생: {rename_e}")
                            
                if not renamed:
                    print(f"  - 다운로드 완료 확인 실패. (다운로드 대기 시간 조정 필요)")
                    initial_files = get_download_files() 


            except Exception as link_e:
                print(f"  - {i+1}번째 요소 처리 중 오류 발생: {link_e}")
                
        return downloaded_count

    except Exception as e:
        # 이 부분이 Chrome 충돌 오류가 발생하는 부분입니다.
        print(f"❌ 다운로드 과정 중 치명적인 오류 발생 (드라이버 충돌 예상): {e}")
        return 0

# --------------------------
# 🌟 메인 실행
# --------------------------

def main():
    # 1. 다운로드 경로 설정 및 생성
    os.makedirs(DOWNLOAD_DIR, exist_ok=True)
    print(f"📁 첨부파일 다운로드 경로: {DOWNLOAD_DIR}")

    # 2. CSV 파일 읽기 및 테스트 데이터 추출
    test_data = None
    try:
        with open(CSV_FILE, 'r', newline='', encoding='utf-8-sig') as f:
            reader = csv.DictReader(f)
            # 🌟 첫 번째 행만 가져와 테스트
            for row in reader:
                test_data = row
                break
    except FileNotFoundError:
        print(f"❌ {CSV_FILE} 파일을 찾을 수 없습니다. 크롤링을 먼저 실행하여 파일을 생성하세요.")
        return
    except Exception as e:
        print(f"❌ CSV 파일 읽기 오류: {e}")
        return

    if not test_data:
        print("❌ CSV 파일이 비어 있어 테스트를 진행할 수 없습니다.")
        return
    
    # 테스트에 사용할 데이터
    url = test_data.get('URL')
    date = test_data.get('등록일')
    title = test_data.get('제목')
    
    if not url or not date or not title:
        print("❌ CSV 데이터가 불완전합니다 (URL, 등록일, 제목 중 하나가 없음).")
        return

    print(f"\n--- 🧪 게시물 1개 다운로드 테스트 시작 ---")
    print(f"URL: {url}")
    print(f"제목: {title}")
    
    # 3. 드라이버 설정
    driver = setup_driver(DOWNLOAD_DIR)
    if not driver:
        return

    try:
        # 4. 첨부파일 다운로드 실행
        downloaded_count = download_attachments_from_page(driver, url, date, title, DOWNLOAD_DIR)

        if downloaded_count > 0:
            print(f"\n✅ 테스트 완료! 총 {downloaded_count}개 파일 다운로드 및 이름 변경 성공.")
        else:
            print(f"\n❌ 테스트 완료. 다운로드된 파일이 없거나 이름 변경에 실패했습니다.")

    finally:
        # 5. 브라우저 종료
        time.sleep(2)
        driver.quit()
        print("\n브라우저 종료")

if __name__ == '__main__':
    main()
```

### 1차 다운로드

```python
import os
import csv
import time
import re
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from glob import glob

# --------------------------
# 🌟 설정 변수
# --------------------------
CSV_FILE = "sh_crw.csv"
DOWNLOAD_DIR = os.path.abspath("./downloads_sh")  # 다운로드 경로
WAIT_TIME = 20
DOWNLOAD_MAX_WAIT_TIME = 15  # 다운로드 완료 대기 시간
RENAME_MAX_RETRIES = 3
NAVIGATION_MAX_RETRIES = 3

# --------------------------
# 🌟 유틸리티 함수
# --------------------------
def get_all_files(download_path):
    """다운로드 폴더 내 모든 파일 경로 반환"""
    return set(glob(os.path.join(download_path, '*')))

def sanitize_filename(name):
    """파일명 특수문자 제거 및 길이 제한"""
    # 괄호와 숫자 시퀀스(예: (1), (2) 등)는 파일명 중복 처리 과정에서 붙는 것이므로,
    # 여기서 최종 파일명에 반영되는 특수문자만 제거합니다.
    name = re.sub(r'[\\/:*?"<>|]', '', name)
    return name[:100].strip() # 길이 제한을 50자에서 100자로 확장하여 안정성 확보

# --------------------------
# 🌟 Selenium 설정
# --------------------------
def setup_driver(download_path):
    chrome_options = Options()
    prefs = {
        "download.default_directory": download_path,
        "download.prompt_for_download": False,
        "download.directory_upgrade": True,
        "plugins.always_open_pdf_externally": True
    }
    chrome_options.add_experimental_option("prefs", prefs)
    chrome_options.add_argument('--disable-gpu')
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    chrome_options.add_argument("--start-maximized")
    # Headless 모드 제거 (사용자 코드 반영)
    # chrome_options.add_argument('--headless')

    try:
        service = Service(ChromeDriverManager().install())
        driver = webdriver.Chrome(service=service, options=chrome_options)
        return driver
    except Exception as e:
        print(f"❌ Chrome 드라이버 설정 오류: {e}")
        return None

# --------------------------
# 🌟 다운로드 함수
# --------------------------
def download_attachments_from_page(driver, url, date, title, download_path):
    downloaded_count = 0
    extensions = ['.pdf', '.xls', '.xlsx', '.txt', '.hwp', '.hwpx', '.doc', '.docx', '.zip']
    ext_regex = r'\.(' + '|'.join([ext[1:] for ext in extensions]) + r')$'

    try:
        print(f"  ➡️ 상세 페이지 접속: {url}")

        # 페이지 접속 재시도
        navigated = False
        for attempt in range(NAVIGATION_MAX_RETRIES):
            try:
                driver.get(url)
                WebDriverWait(driver, WAIT_TIME).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, 'div.pageContainer, #mainform'))
                )
                navigated = True
                break
            except Exception as e:
                if attempt < NAVIGATION_MAX_RETRIES - 1:
                    print(f"    - ⚠️ 접속 실패, 10초 후 재시도 ({attempt+1})")
                    time.sleep(10)
                else:
                    print(f"❌ 페이지 접속 실패: {e}")
                    return 0
        if not navigated:
            return 0

        # 파일 요소 조회
        xpath_query = (
            "//*[self::a][contains(@onclick, 'FileDownLoad') "
            "or contains(text(), '.pdf') or contains(text(), '.xls') "
            "or contains(text(), '.hwp') or contains(text(), '.zip')]"
        )

        file_elements = driver.find_elements(By.XPATH, xpath_query)
        if not file_elements:
            return 0

        # 첨부파일 반복 다운로드
        for i in range(len(file_elements)):
            # DOM 재조회 (Stale Element Reference 방지)
            file_elements = driver.find_elements(By.XPATH, xpath_query)
            if i >= len(file_elements):
                break
                
            element = file_elements[i]
            
            try:
                original_text = element.text.strip()
                match = re.search(ext_regex, original_text, re.IGNORECASE)
                ext_from_text = match.group(0).lower() if match else '' # 원본 텍스트에서 추출한 확장자

                # 클릭 전 기존 파일 목록
                initial_files = get_all_files(download_path)

                # 클릭 → 다운로드 시작
                print(f"    - 다운로드 시도: '{original_text}'")
                driver.execute_script("arguments[0].click();", element)
                time.sleep(3)  # JS 처리 및 서버 생성 지연 대응

                # 다운로드 완료 대기
                start_time = time.time()
                downloaded_file = None
                while time.time() - start_time < DOWNLOAD_MAX_WAIT_TIME:
                    time.sleep(1)
                    current_files = get_all_files(download_path)
                    new_files = current_files - initial_files
                    
                    for f_path in new_files:
                        # 다운로드 완료된 파일 (임시 확장자가 없는 경우)
                        if not f_path.endswith(('.crdownload', '.tmp', '.part')):
                            downloaded_file = f_path
                            break
                    if downloaded_file:
                        break

                if not downloaded_file:
                    print(f"    - ❌ {DOWNLOAD_MAX_WAIT_TIME}초 내 다운로드 완료 확인 실패. (원본 텍스트: {original_text})")
                    continue

                # 🌟🌟🌟 파일명 변경 로직 수정: 원본 텍스트 기반으로 최종 파일명 구성 🌟🌟🌟

                # 1. 원본 텍스트에서 제목 부분만 추출 (확장자 제거)
                clean_title_base = re.sub(ext_regex, '', original_text, flags=re.IGNORECASE)
                clean_title = sanitize_filename(clean_title_base)

                # 2. 최종 확장자 결정: 원본 텍스트에 확장자가 있으면 그것을 사용, 없으면 다운로드된 파일의 실제 확장자를 사용
                if ext_from_text:
                    final_ext = ext_from_text
                else:
                    final_ext = os.path.splitext(os.path.basename(downloaded_file))[1]
                    
                # 3. 최종 파일명 구성
                target_name = f"({date}) {clean_title}{final_ext}"

                # 4. 파일명 충돌 처리
                final_path = os.path.join(download_path, target_name)
                collision = 1
                while os.path.exists(final_path):
                    base, ext_name_tmp = os.path.splitext(target_name)
                    # 파일명에 기존 충돌 번호가 있다면 제거하고 새로 붙여야 합니다.
                    base_no_collision = re.sub(r'(\s*\(\d+\))?$', '', base)
                    final_path = os.path.join(download_path, f"{base_no_collision}({collision}){ext_name_tmp}")
                    collision += 1

                # 5. 파일명 변경 시도
                for attempt in range(RENAME_MAX_RETRIES):
                    try:
                        os.rename(downloaded_file, final_path)
                        print(f"    - 다운로드 및 이름 변경 완료: {os.path.basename(final_path)}")
                        downloaded_count += 1
                        break
                    except PermissionError:
                        time.sleep(1)
                    except Exception as e:
                        print(f"    - ❌ 파일명 변경 실패: {e}")
                        break

            except Exception as e:
                print(f"    - 다운로드 처리 오류: {e}")
                continue

        return downloaded_count

    except Exception as e:
        print(f"❌ 다운로드 오류: {e}")
        return 0

# --------------------------
# 🌟 메인 실행
# --------------------------
def main():
    os.makedirs(DOWNLOAD_DIR, exist_ok=True)
    print(f"📁 다운로드 경로: {DOWNLOAD_DIR}")

    try:
        # utf-8-sig로 인코딩을 지정하여 BOM 문제를 해결합니다.
        with open(CSV_FILE, 'r', newline='', encoding='utf-8-sig') as f: 
            reader = csv.DictReader(f)
            all_data = list(reader)
    except Exception as e:
        print(f"❌ CSV 파일 읽기 오류: {e}")
        return

    total_records = len(all_data)
    total_downloaded = 0
    print(f"\n--- 🚀 총 {total_records}개의 게시물 다운로드 시작 ---")

    driver = setup_driver(DOWNLOAD_DIR)
    if not driver:
        return

    try:
        for idx, data in enumerate(all_data):
            url = data.get('URL')
            date = data.get('등록일')
            title = data.get('제목')

            if not url or not date or not title:
                print(f"\n--- [{idx+1}/{total_records}] 데이터 불완전, 건너뜀 ---")
                continue

            print(f"\n--- [{idx+1}/{total_records}] {date} - {title} 처리 ---")
            downloaded_count = download_attachments_from_page(driver, url, date, title, DOWNLOAD_DIR)
            total_downloaded += downloaded_count

    finally:
        time.sleep(2)
        driver.quit()
        print("\n브라우저 종료")

    print(f"\n=======================================================")
    print(f"🎉 완료! 총 {total_records}개 게시물에서 {total_downloaded}개 파일 다운로드 성공.")
    print(f"저장 경로: {DOWNLOAD_DIR}")
    print(f"=======================================================")

if __name__ == '__main__':
    main()
```

### 2차 다운로드 (다운로드 안된 것들 재시도)

```python
import os
import csv
import time
import re
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from glob import glob


# -----------------------------
# 🌟 설정값
# -----------------------------
CSV_FILE = "sh_notice.csv"
DOWNLOAD_DIR = os.path.abspath("./downloads_sh")
WAIT_TIME = 20
DOWNLOAD_MAX_WAIT_TIME = 30  # ★ 150초 → 30초로 변경
RENAME_MAX_RETRIES = 3
NAVIGATION_MAX_RETRIES = 3


# -----------------------------
# 🌟 유틸 함수
# -----------------------------
def get_all_files(download_path):
    return set(glob(os.path.join(download_path, '*')))


def sanitize_filename(name):
    name = re.sub(r'[\\/:*?"<>|]', '', name)
    return name[:100].strip()


# -----------------------------
# 🌟 Selenium 설정
# -----------------------------
def setup_driver(download_path):
    chrome_options = Options()
    prefs = {
        "download.default_directory": download_path,
        "download.prompt_for_download": False,
        "download.directory_upgrade": True,
        "plugins.always_open_pdf_externally": True
    }
    chrome_options.add_experimental_option("prefs", prefs)
    chrome_options.add_argument("--start-maximized")
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")

    try:
        service = Service(ChromeDriverManager().install())
        driver = webdriver.Chrome(service=service, options=chrome_options)
        return driver
    except Exception as e:
        print(f"❌ Chrome 드라이버 설정 오류: {e}")
        return None


# -----------------------------
# 🌟 첨부파일 다운로드 함수
# -----------------------------
def download_attachments_from_page(driver, url, date, title, download_path):
    downloaded_count = 0
    extensions = ['.pdf', '.xls', '.xlsx', '.txt', '.hwp', '.hwpx', '.doc', '.docx', '.zip']
    ext_regex = r'\.(' + '|'.join(ext[1:] for ext in extensions) + r')$'

    print(f"  ➡ 상세 페이지 접속: {url}")

    # 접속 재시도
    for attempt in range(NAVIGATION_MAX_RETRIES):
        try:
            driver.get(url)
            WebDriverWait(driver, WAIT_TIME).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "div.pageContainer, #mainform"))
            )
            break
        except Exception:
            if attempt == NAVIGATION_MAX_RETRIES - 1:
                print("❌ 페이지 접속 실패, 건너뜀")
                return 0
            time.sleep(5)

    # 첨부파일 요소 탐색
    xpath_query = (
        "//*[self::a][contains(@onclick, 'FileDownLoad') "
        "or contains(text(), '.pdf') or contains(text(), '.xls') "
        "or contains(text(), '.hwp') or contains(text(), '.zip')]"
    )

    file_elements = driver.find_elements(By.XPATH, xpath_query)
    if not file_elements:
        return 0

    # 파일 반복 처리
    for i in range(len(file_elements)):
        file_elements = driver.find_elements(By.XPATH, xpath_query)
        if i >= len(file_elements):
            break

        element = file_elements[i]
        original_text = element.text.strip()

        # 확장자 추출
        match = re.search(ext_regex, original_text, re.IGNORECASE)
        ext_from_text = match.group(0).lower() if match else ""

        # 최종 파일명 생성
        clean_title_base = re.sub(ext_regex, "", original_text, flags=re.IGNORECASE)
        clean_title = sanitize_filename(clean_title_base)
        final_ext = ext_from_text if ext_from_text else ".pdf"

        expected_name = f"({date}) {clean_title}{final_ext}"
        expected_path = os.path.join(download_path, expected_name)

        # -------------------------
        # 🌟 이미 존재하는 파일이면 건너뛰기
        # -------------------------
        if os.path.exists(expected_path):
            print(f"    - ⏩ 이미 존재: {expected_name} → 건너뜀")
            continue

        print(f"    - 다운로드 시도: {original_text}")

        initial_files = get_all_files(download_path)

        try:
            driver.execute_script("arguments[0].click();", element)
            time.sleep(2)
        except Exception as e:
            print(f"    - ❌ 클릭 실패: {e}")
            continue

        # 다운로드 완료 대기 (최대 DOWNLOAD_MAX_WAIT_TIME)
        start = time.time()
        downloaded_file = None

        while time.time() - start < DOWNLOAD_MAX_WAIT_TIME:
            current_files = get_all_files(download_path)
            new_files = current_files - initial_files

            for f in new_files:
                if not f.endswith((".crdownload", ".tmp", ".part")):
                    downloaded_file = f
                    break

            if downloaded_file:
                break
            time.sleep(1)

        # 실패 처리
        if not downloaded_file:
            print(f"    - ❌ 다운로드 실패 (30초 초과): {original_text}")
            continue

        # 최종 경로 충돌 처리
        final_path = expected_path
        if os.path.exists(final_path):
            base, ext = os.path.splitext(expected_name)
            count = 1
            while os.path.exists(final_path):
                final_path = os.path.join(download_path, f"{base}({count}){ext}")
                count += 1

        # 이름 변경 시도
        try:
            os.rename(downloaded_file, final_path)
            print(f"    - 다운로드 완료: {os.path.basename(final_path)}")
            downloaded_count += 1
        except Exception as e:
            print(f"    - ❌ 파일명 변경 실패: {e}")
            continue

    return downloaded_count


# -----------------------------
# 🌟 메인
# -----------------------------
def main():
    os.makedirs(DOWNLOAD_DIR, exist_ok=True)
    print(f"📁 다운로드 경로: {DOWNLOAD_DIR}")

    try:
        with open(CSV_FILE, "r", encoding="utf-8-sig") as f:
            all_data = list(csv.DictReader(f))
    except Exception as e:
        print(f"❌ CSV 읽기 오류: {e}")
        return

    driver = setup_driver(DOWNLOAD_DIR)
    if not driver:
        return

    total_records = len(all_data)
    total_downloaded = 0

    print(f"\n--- 🚀 총 {total_records}개의 게시물 다운로드 시작 ---")

    try:
        for idx, data in enumerate(all_data):
            url = data["URL"]
            date = data["등록일"]
            title = data["제목"]

            print(f"\n--- [{idx+1}/{total_records}] {date} - {title} ---")

            total_downloaded += download_attachments_from_page(
                driver, url, date, title, DOWNLOAD_DIR
            )

    finally:
        driver.quit()
        print("\n브라우저 종료")

    print("\n====================================")
    print(f"🎉 완료! 총 {total_downloaded}개 파일 다운로드 성공")
    print("====================================")


if __name__ == "__main__":
    main()
```
