```python
artists = [("Walt Disney","classic animation style, early Mickey Mouse era, 1928, black and white, Steamboat Willie aesthetic, vintage cartoon")]

# artists로 받은 데이터를 바탕으로 4개의 이미지를 생성, 출력, 저장하기
for i in range(4):
    artist, subject = artists[0]  # artists 리스트에서 첫 번째 항목 사용
    print(i, f'[{artist}], subject: {subject}')
    prompt = f'{artist}, {subject}'

    response = client.images.generate(
        model="dall-e-3",
        prompt=prompt,
        size="1024x1024",
        quality="standard",
        style="vivid"
    )
    print(f'{artist} : {response.data[0].url}')
    url = response.data[0].url
    display.display(display.Image(url=url, width=300, height=300))
    time.sleep(10)

    # 생성된 이미지 저장
    import requests
    # 이미지 저장 경로 설정
    save_path = f'{artist}_{i}.png'
    # 이미지 저장
    img_data = requests.get(url).content
    with open(save_path, 'wb') as f:
        f.write(img_data)
    print(f'{save_path} 저장 완료')
```

```python
# 구글 코랩 환경에서 DALL-E 3로 이미지를 생성하고, 바로 보여주고, 다운로드 링크를 제공하는 예시 코드입니다.
import time
import requests
from IPython.display import display, Image
from google.colab import files

artist = "Walt Disney"
subject = "classic animation style, early Mickey Mouse era, 1928, black and white, Steamboat Willie aesthetic, vintage cartoon"
prompt = f'{artist}, {subject}'

response = client.images.generate(
    model="dall-e-3",
    prompt=prompt,
    size="1024x1024",
    quality="standard",
    style="vivid"
)
url = response.data[0].url

# 이미지 바로 출력
display(Image(url=url, width=300, height=300))

# 이미지 파일 자동 생성 및 다운로드 링크 제공
filename = f"{artist.replace(' ', '_')}_dall-e3.png"
img_data = requests.get(url).content
with open(filename, 'wb') as f:
    f.write(img_data)

print(f"이미지가 저장되었습니다: {filename}")
files.download(filename)
```
