```python
# 디바이스 설정
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
# 모델 로드
from diffusers import StableDiffusionPipeline

model_name = 'runwayml/stable-diffusion-v1-5'
pipe = StableDiffusionPipeline.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    use_safetensors=True    
).to(device)
```

```python
# seed 설정
def set_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
```

### 이미지 생성

```python
# 단일 이미지 생성
set_seed(42)

prompt = 'asian gen-z girl, Celestial map of paradise, 3D render, solarpunk, bright lighting, art deco — v 5'

result = pipe(
    prompt,
    num_inference_steps=50,
    guidance_scale=7.5,
)
display(result.images[0])   # PIL 이미지 객체
```

```python
prompt = 'A serence landscape of mountains during sunrise, oil painting'
```

```python
prompts = [
    "A futuristic city skyline at sunset",
    "A portrait of a medieval knight in armor",
    "A colorful abstract painting with geometric shapes"
]

result = pipe(
    prompts,
    num_inference_steps=50,
    guidance_scale=7.5,       # 생성 이미지의 퀄리티 조절, 보통 7.5~10 사이
    width=768,
    height=512
)

print(result.images) # 이미지 객체 리스트
```

```python
# 배치 결과 시각화
import matplotlib.pyplot as plt

images = result.images
n = len(images)

plt.figure(figsize=(4, 3 *n))
for i, img in enumerate(images):
    plt.subplot(1, n, i+1)
    plt.imshow(img)
    plt.title(prompts[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
for idx, img in enumerate(images):
    filename = f'stable_diffusion_{idx}.png'
    img.save(filename)
    print(f'Saved: {filename}')
```
