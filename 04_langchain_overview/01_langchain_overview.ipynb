## Langchain이란?  

Langchain은 대규모 언어 모델(LLM)과 외부 데이터, 다양한 도구들을 쉽게 연결하여 유용한 AI 응용 프로그램을 만들 수 있게 도와주는 파이썬 오픈소스 프레임워크입니다.
Langchain을 활용하면 단순한 질문-답변 시스템을 넘어서, 문서 검색, 대화형 챗봇, 복잡한 워크플로우 자동화 등 다양한 AI 서비스와 기능들을 개발할 수 있습니다.

### Langchain의 주요 구성 요소

- **Model I/O (모델 입출력)**  
  다양한 언어 모델(OpenAI GPT, HuggingFace 등)과 연결해 텍스트 생성·분석 등의 작업을 손쉽게 할 수 있습니다.

- **Retrieval (정보 검색)**  
  외부 문서, 데이터베이스 등의 정보 소스를 연결하여, LLM이 실시간으로 데이터에 접근해 더 정확한 답변이나 정보를 제공할 수 있습니다.  
  (예: 사내 문서 검색, FAQ 자동화 등)

- **Memory (메모리/기억)**  
  대화의 맥락이나 이전 이력을 저장하고 활용해서, 자연스러운 대화나 연속성 있는 작업 처리가 가능합니다.

- **Chains (체인/연쇄 작업)**  
  여러 LLM 작업, 데이터 처리, 외부 API 호출 등을 단계별로 연결하여, 복잡한 논리 흐름이나 조건식을 가진 AI 프로세스를 설계할 수 있습니다.

- **Agents/Tools (에이전트/도구)**  
  다양한 외부 도구(검색엔진, 계산기, 데이터API 등)와 연동하거나, AI가 목적에 맞게 여러 도구를 스스로 선택·활용하도록 설계할 수 있습니다.

---

### Langchain을 학습해야 하는 이유

- LLM의 강점을 실제 서비스와 제품에 쉽게 적용
- 외부 데이터와 결합해 정확성·신뢰성 높은 AI 구축 가능
- Open-source 기반으로 다양한 예시와 생태계가 활발함

따라서, Langchain을 익히면 실전 AI 서비스를 빠르게 구축하고, 새로운 LLM 활용 방식을 창의적으로 시도할 수 있습니다.

## LangChain 패키지 구조

LangChain은 모듈화된 구조로 여러 패키지로 나뉘어져 있습니다. 각 패키지는 특정 목적과 역할을 가지고 있습니다.

### 📦 주요 패키지들

#### 1. `langchain` (메인 통합 패키지)
- **역할**: 모든 LangChain 기능을 사용할 수 있는 통합 패키지
- **특징**: 
  - 가장 일반적으로 설치하는 패키지
  - 다른 하위 패키지들을 함께 포함
  - 빠르게 시작하기에 적합
- **설치**: `pip install langchain`

---

#### 2. `langchain_core` (핵심 추상화 레이어)
- **역할**: LangChain의 **핵심 인터페이스와 추상 클래스** 제공
- **특징**:
  - 모든 LangChain 패키지의 **기반**이 되는 코어 라이브러리
  - 의존성이 매우 적고 가벼움
  - 주요 추상화 클래스들:
    - `Runnable`: 실행 가능한 모든 컴포넌트의 기본 인터페이스
    - `BaseRetriever`: 검색 시스템의 기본 클래스
    - `BaseMemory`: 메모리/대화 이력 관리 기본 클래스
    - `BaseChatModel`, `BaseLLM`: LLM 관련 기본 클래스
- **사용 예시**: 커스텀 컴포넌트를 만들 때 이 패키지의 추상 클래스를 상속받아 구현

---

#### 3. `langchain_community` (커뮤니티 통합)
- **역할**: **외부 서비스 및 도구**와의 통합 모듈 제공
- **특징**:
  - 서드파티 서비스 연동 (벡터DB, LLM 제공업체, 도구 등)
  - 커뮤니티에서 기여한 통합 모듈들
  - 주요 통합 예시:
    - 벡터 데이터베이스: `Chroma`, `FAISS`, `Pinecone`, `Weaviate`
    - LLM: `HuggingFace`, `Cohere`, `Anthropic`
    - 문서 로더: `PDF`, `CSV`, `HTML` 등
    - 도구: `Google Search`, `Wikipedia`, `Calculator` 등
- **설치**: `pip install langchain-community`
- **주의**: 각 통합마다 추가 의존성이 필요할 수 있음

---

#### 4. `langchain_experimental` (실험적 기능)
- **역할**: **실험적이고 베타 단계**의 새로운 기능 제공
- **특징**:
  - 아직 안정화되지 않은 혁신적인 기능들
  - API가 변경될 수 있음
  - 예시:
    - 자율 에이전트 (Autonomous Agents)
    - 복잡한 체인 패턴
    - 새로운 프롬프트 기법
- **주의**: 
  - 프로덕션 환경에서는 신중하게 사용
  - 버전 업데이트 시 breaking change 가능성

---

#### 5. `langchain_text_splitters` (텍스트 분할)
- **역할**: **문서를 작은 청크(chunk)로 분할**하는 기능 제공
- **특징**:
  - RAG(Retrieval-Augmented Generation) 시스템에서 필수적
  - 긴 문서를 임베딩하기 적절한 크기로 나누기
  - 주요 클래스:
    - `RecursiveCharacterTextSplitter`: 재귀적으로 문자 기준 분할
    - `CharacterTextSplitter`: 단순 문자 기준 분할
    - `TokenTextSplitter`: 토큰 수 기준 분할
    - `MarkdownTextSplitter`: 마크다운 구조 기준 분할
- **설치**: `pip install langchain-text-splitters`
- **사용 시나리오**: 
  - PDF, 긴 문서를 벡터DB에 저장하기 전 처리
  - 토큰 제한이 있는 LLM에 입력하기 위한 분할

---

### 📊 패키지 구조 요약

```
langchain (통합 패키지)
├── langchain_core (핵심 추상화)          ← 모든 것의 기반
├── langchain_community (외부 통합)       ← Chroma, FAISS, HuggingFace 등
├── langchain_experimental (실험 기능)    ← 새로운 아이디어, 베타 기능
└── langchain_text_splitters (문서 분할) ← RAG를 위한 청킹
```

---

### 💡 실무에서 어떻게 사용하나?

**초보자/빠른 프로토타입**:
```python
# 간단하게 메인 패키지만 설치
pip install langchain
```

**선택적 설치 (프로덕션)**:
```python
# 필요한 것만 골라서 설치 (의존성 최소화)
pip install langchain-core
pip install langchain-community  # Chroma, FAISS 등 사용 시
pip install langchain-text-splitters  # RAG 시스템 구축 시
pip install langchain-openai  # OpenAI 사용 시
```
