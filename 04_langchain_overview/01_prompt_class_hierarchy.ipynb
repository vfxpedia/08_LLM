# langchain prompt class hierarchy

LangChain의 프롬프트(Prompt) 관련 클래스 계층 구조는 다양한 프롬프트 생성 요구사항을 충족하기 위해 설계되어 있습니다. 계층 구조의 최상위에는 추상 클래스인 `BasePromptTemplate`이 있으며, 실제로는 다양한 하위 클래스들이 주요 역할을 담당합니다.[^2_2][^2_5][^2_6]

### LangChain Prompt Class Hierarchy

- **BasePromptTemplate**
    - 모든 프롬프트 템플릿의 추상 베이스 클래스입니다. 입력 변수, 포매팅, 출력 파서 등 핵심 메서드를 정의합니다.[^2_5][^2_6]
- **PipelinePromptTemplate**
    - 여러 프롬프트 템플릿을 조합하여 파이프라인 형태로 구성할 수 있는 클래스입니다.[^2_6]
- **StringPromptTemplate**
    - 문자열 기반 프롬프트 템플릿의 베이스 클래스입니다.
        - → **PromptTemplate**: 표준 문자열 프롬프트 생성에 사용.[^2_2][^2_6]
        - → **FewShotPromptTemplate**: Few-shot 예시를 포함하는 프롬프트에 사용.
        - → **FewShotPromptWithTemplates**: 템플릿별 few-shot 예시 제공.[^2_6]
- **BaseChatPromptTemplate**
    - 챗 기반 프롬프트 템플릿의 베이스 클래스입니다.
        - → **ChatPromptTemplate**: 여러 메시지(role별)를 포함한 챗 프롬프트.[^2_8][^2_6]
        - → **AgentScratchPadChatPromptTemplate**: 에이전트 활용 챗 프롬프트.
- **BaseMessagePromptTemplate**
    - 메시지 단위 템플릿의 베이스 클래스입니다.
        - → **MessagesPlaceholder**: 메시지 리스트를 그대로 전달.[^2_6]
        - → **ChatMessagePromptTemplate**, **HumanMessagePromptTemplate**, **AIMessagePromptTemplate**, **SystemMessagePromptTemplate**: 각 역할별 메시지를 위한 템플릿 클래스.[^2_8][^2_6]
        - → **BaseStringMessagePromptTemplate**: 문자열 프롬프트 기반 메시지 템플릿의 베이스 클래스.
- **ImagePromptTemplate**
    - 멀티모달(이미지+텍스트) 모델에서 이미지 프롬프트 생성.[^2_6]


### 예시 다이어그램 (텍스트)

```
BasePromptTemplate
├── PipelinePromptTemplate
├── StringPromptTemplate
│   ├── PromptTemplate
│   ├── FewShotPromptTemplate
│   └── FewShotPromptWithTemplates
├── BaseChatPromptTemplate
│   ├── ChatPromptTemplate
│   └── AgentScratchPadChatPromptTemplate
├── BaseMessagePromptTemplate
│   ├── MessagesPlaceholder
│   ├── ChatMessagePromptTemplate
│   ├── HumanMessagePromptTemplate
│   ├── AIMessagePromptTemplate
│   └── SystemMessagePromptTemplate
├── BaseStringMessagePromptTemplate
└── ImagePromptTemplate
```

이 계층 구조는 텍스트, 챗, 이미지 등 다양한 입력 방식에 맞춰 확장되어 있으며, 조합과 커스터마이징이 자유롭습니다. 실제 다이어그램이나 UML을 원한다면 추가 안내가 가능합니다.[^2_2][^2_8][^2_6]
